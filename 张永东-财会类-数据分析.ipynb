```python
import xlrd
paths = r"D:\Desktop\新建文件夹\财会.xlsx"
def rtSheetList(path):
    workbook99 = xlrd.open_workbook(path)
    sheetList = workbook99.sheet_names()
    return sheetList, workbook99


def rtSheetDate(workbook99, sheetName):
    lists99 = []
    sheet29 = workbook99.sheet_by_name(sheetName)
    for i in range(sheet29.nrows):
        lists99.append(sheet29.row_values(i))
    return lists99

data_sheet_name = rtSheetList(paths)
print(data_sheet_name)

data_sheet_dct = {}
for i in data_sheet_name[0]:
    data_list = rtSheetDate(data_sheet_name[1],i)
    data_sheet_dct[i] = len(data_list)
    for j in range(len(data_list)):
        print(data_list[j])

def get_index_list(lists,data):
    for i in range(len(lists)):
        if data == str(lists[i]):
            return i
    return -1
```

```python
data_list2 =rtSheetDate(data_sheet_name[1],data_sheet_name[0][0])
dct = {}
test1 = []
for i in data_list2:
    row_data = i[0]
    if not row_data in test1:
        test1.append(row_data)
        dct[row_data] = 0
    dct[row_data] += 1
#     print(i)

dct2 = {}
bjs = 0
for k,v in dct.items():
    bjs +=1
    if bjs == 20:
        break
    dct2[k]=v
    


    
# 导入模块
from pyecharts import options as opts
from pyecharts.charts import Pie
#准备数据
label=[k for k,v in dct2.items()]  
values = [v for k,v in dct2.items()]
# 自定义函数
def pie_base():
    c = (
        Pie()
        .add("",[list(z) for z in zip(label,values)])
        .set_global_opts(title_opts = opts.TitleOpts(title="职业占比"))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c} {d}%"))   # 值得一提的是，{d}%为百分比
    )
    return c
# 调用自定义函数生成render.html
pie_base().render_notebook()
# pie_base().render()
```
